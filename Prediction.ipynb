## Remove duplicates

### Subtask:
Identify and remove duplicate rows from the DataFrame.


**Reasoning**:
Identify and remove duplicate rows from the DataFrame, then verify the removal.



```python
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows found: {num_duplicates}")
df.drop_duplicates(inplace=True)
num_duplicates_after_removal = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {num_duplicates_after_removal}")
```

## Check missing values
